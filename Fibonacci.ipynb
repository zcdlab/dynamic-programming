# Fibonacci Probelm

# 1. Rescursive

```python
def fib(n):
    result = 0
    
    if(n == 1 or n == 2):
        result = 1
    else:
        result = fib(n-2) + fib(n-1)
        
    return result
```

```python
> You will see the computational cost exponentially increases.
```

```python
print(fib(1))
%timeit fib(1)
```

```python
print(fib(5))
%timeit fib(5)
```

```python
print(fib(10))
%timeit fib(10)
```

```python
print(fib(30))
%timeit fib(30)
```

```python
# No need to calculate the time, since it is obviously expensive.
print(fib(40))
```

# Memory Status

```python
def fib(n, memory_array):
    if(memory_array[n] is not None):
        return memory_array[n]
    else:
        result = 0
        if(n == 1 or n == 2):
            result = 1
        else:
            result = fib(n - 2, memory_array) + fib(n - 1, memory_array)
    
    memory_array[n] = result
    return result
```

```python
memory_array = [None] * (10000)
print(fib(1, memory_array))
%timeit fib(1, memory_array)
```

```python
memory_array = [None] * (10000)
print(fib(5, memory_array))
%timeit fib(5, memory_array)
```

# Bottom Up

```python
def fib(n):
    result = 0
    array = [None] * (n + 1)
    if(n == 1 or n == 2):
        result = 1
    else:
        array[1] = 1
        array[2] = 1
        for i in range(3, n + 1):
            array[i] = array[i - 2] + array[i - 1]
        
        result = array[n]
        
    return result 
```

```python
print(fib(1))
%timeit fib(1)
```

```python
print(fib(5))
%timeit fib(5)
```

```python
print(fib(10))
%timeit fib(10)
```

```python
print(fib(50))
%timeit fib(50)
```
